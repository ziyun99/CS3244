```python
import os
import numpy as np
import shutil
import random
root_dir = "/content/drive/MyDrive/y3s2/CS3244 ML Project/CS3244/our_dataset/"# data root path
new_dir = "/content/drive/MyDrive/y3s2/CS3244 ML Project/CS3244/data/"
classes_dir = [ 'fabric',
                'cardboard',
                'glass',
                'metal',
                'paper',
                'plastic',
                'residual',
                'compost',
                'hazardous'] #total labels

# val_ratio = 0.15
test_ratio = 0.20

for cls in classes_dir:
    os.makedirs(new_dir +'train/' + cls)
    # os.makedirs(root_dir +'val/' + cls)
    os.makedirs(new_dir +'test/' + cls)


    # Creating partitions of the data after shuffeling
    src = root_dir + cls # Folder to copy images from

    allFileNames = os.listdir(src)
    np.random.shuffle(allFileNames)
    train_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                              [int(len(allFileNames)* (1 - test_ratio))])

    print(src)

    train_FileNames = [src+'/'+ name for name in train_FileNames.tolist()]
    # val_FileNames = [src+'/' + name for name in val_FileNames.tolist()]
    test_FileNames = [src+'/' + name for name in test_FileNames.tolist()]

    print('Total images: ', len(allFileNames))
    print('Training: ', len(train_FileNames))
    # print('Validation: ', len(val_FileNames))
    print('Testing: ', len(test_FileNames))

    # Copy-pasting images
    for name in train_FileNames:
        shutil.copy(name, new_dir +'train/' + cls)

    # for name in val_FileNames:
    #     shutil.copy(name, new_dir +'val/' + cls)

    for name in test_FileNames:
        shutil.copy(name, new_dir +'test/' + cls)
```
