```python
from PIL import Image
import os
 
folder = "/content/drive/MyDrive/y3s2/CS3244 ML Project/CS3244/Dataset"
w = 224
h = 224
for root, dirs, files in os.walk(folder):
  # print(f)
  for name in files:
      fname = os.path.join(root, name)
      exts = ['.jpg', '.jpeg','.png']
      for ext in exts:
        if ext in fname:
          # print(fname)
          image = Image.open(fname)
          # image.show()
          # print(image.size)
          # display(image)
          width, height = image.size 
          if width > height:
           image.thumbnail((image.size[0], 224), Image.ANTIALIAS)
          else:
           image.thumbnail((224, image.size[1]), Image.ANTIALIAS)
          # print(image.size)
          # display(image)

          width, height = image.size   # Get dimensions

          left = (width - 224)/2
          top = (height - 224)/2
          right = (width + 224)/2
          bottom = (height + 224)/2

          # Crop the center of the image
          if width > 224 and height > 224:
            image = image.crop((left, top, right, bottom))
          else:
            image =image.resize((224, 224), Image.ANTIALIAS)

          print(image.size)
          image = image.convert('RGB')
          # display(image)

          image.save(fname, "JPEG")
```

```python

print("Shrink images in the folder")
folder = input("folder: ")
wtop = int(input("wtop: "))
htop = int(input("htop: "))
wdown = int(input("wdown: "))
hdown = int(input("hdown: "))
for i in os.listdir(folder):
    file = f"{folder}\\{i}"
    im = Image.open(file)
    im = im.crop((wtop, htop, wdown, hdown))
    im.save(file)
```
